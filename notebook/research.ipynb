```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv('books.csv', sep = ';', error_bad_lines = False, encoding = 'latin-1')
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
```

```python
books.head()
```

```python
books.rename(columns = {"Book-Title":"title",
                        "Book-Author":"author",
                        "Year-Of-Publication":"year",
                        "Publisher":"publisher",
                        "Image-URL-L":"image_url"}, inplace = True)
```

```python
books.head()
```

```python
users = pd.read_csv('users.csv', sep = ';', error_bad_lines = False, encoding = 'latin-1')
```

```python
users.head()
```

```python
users.shape
```

```python
users.rename(columns = {"User-ID":"user_id", "Location":"location", "Age":"age"}, inplace=True)
```

```python
users.head()
```

```python
ratings = pd.read_csv('ratings.csv', sep = ';', error_bad_lines = False, encoding = 'latin-1')
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns = {"User-ID":"user_id", "Book-Rating":"rating"}, inplace=True)
```

```python
ratings.head()
```

```python
ratings["user_id"].value_counts()
```

```python
x = ratings["user_id"].value_counts() > 250
```

```python
print(x)
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings["user_id"].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on="ISBN")
```

```python
ratings_with_books.head()
```

```python
number_ratings = ratings_with_books.groupby("title")["rating"].count().reset_index()
```

```python
number_ratings.head()
```

```python
number_ratings.rename(columns = {"rating":"no_of_ratings"}, inplace=True)
```

```python
number_ratings.head()
```

```python
final_rating = ratings_with_books.merge(number_ratings, on="title")
```

```python
final_rating.head()
```

```python
final_rating = final_rating[final_rating['no_of_ratings'] >= 50]
```

```python
final_rating.shape
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
```

```python
final_rating.isnull().sum()
```

```python
#Creating pivot table to cluster data based on ratings
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0, inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
book_pivot.iloc[89,:]
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[89,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(book_pivot.index[suggestion[i]])
```
