```python
import numpy as np
import pandas as pd
```

```python
books = books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Image-URL-L']]
```

```python
books.rename(columns = {"Book-Title":"title",
                        "Book-Author":"author",
                        "Year-Of-Publication":"year",
                        "Publisher":"publisher",
                        "Image-URL-L":"image_url"}, inplace = True)
```

```python
ratings = pd.read_csv('ratings.csv', sep = ';', error_bad_lines = False, encoding = 'latin-1')
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings.rename(columns = {"User-ID":"user_id", "Book-Rating":"rating"}, inplace=True)
```

```python
ratings.head()
```

```python
ratings["user_id"].value_counts()
```

```python
x = ratings["user_id"].value_counts() > 100
```

```python
print(x)
x[x].shape
```

```python
y = x[x].index
```

```python
y
```

```python
ratings = ratings[ratings["user_id"].isin(y)]
```

```python
ratings.head()
```

```python
ratings_with_books = ratings.merge(books, on="ISBN")
```

```python
ratings_with_books.head()
```

```python
number_ratings = ratings_with_books.groupby("title")["rating"].count().reset_index()
```

```python
number_ratings.head()
```

```python
number_ratings.rename(columns = {"rating":"no_of_ratings"}, inplace=True)
```

```python
number_ratings.head()
```

```python
final_rating = ratings_with_books.merge(number_ratings, on="title")
```

```python
final_rating.head()
```

```python
final_rating = final_rating[final_rating['no_of_ratings'] >= 40]
```

```python
final_rating.shape
```

```python
final_rating.head()
```

```python
final_rating.drop_duplicates(['user_id', 'title'], inplace=True)
```
